```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
colors = [['gold', 'olive'],
          ['lawngreen', 'forestgreen']]
n = 100000

p_sierpinski = [1./3, 1./3, 1./3]
p_barnsley = [0.02, 0.09, 0.10, 0.79]
m_sierpinski = [[0.5, 0, 0, 0.5, 0.25, np.sqrt(3.)/4],
                [0.5, 0, 0, 0.5, 0.0, 0],
                [0.5, 0, 0, 0.5, 0.5, 0]]
m_barnsley = np.array([[0.001, 0.0, 0.0, 0.16, 0.0, 0.0],
              [-0.15, 0.28, 0.26, 0.24, 0.0, 0.44],
              [ 0.2,-0.26, 0.23, 0.22, 0.0, 1.6],
              [ 0.85, 0.04,-0.04, 0.85, 0.0, 1.6]])
```

```python
def draw_fractal(n, m_fractal, p_fractal):
    points = []
    x, y = 0, 0
    for _ in range(n):
        matrix = m_fractal[np.random.choice(len(m_fractal), p=p_fractal)]
        x_new = matrix[0] * x + matrix[1] * y + matrix[4]
        y_new = matrix[2] * x + matrix[3] * y + matrix[5]

        points.append([x_new, y_new])
        x, y = x_new, y_new
    return np.array(points)
```

```python
fig, ax = plt.subplots()
fig.suptitle(f'Sierpinski Triangle IFS with n={n}')
sierpinski = draw_fractal(n, m_sierpinski, p_sierpinski)
ax.scatter(sierpinski[:,0], sierpinski[:,1], s=0.006, marker='.', color=colors[0][0])
plt.savefig('../results/plots/03_sierpinski_triangle_ifs', dpi=1200)
plt.show()
plt.close()
```

```python
fig, ax = plt.subplots()
fig.suptitle(f'Barnsley Fern IFS with n={n}')
barnsley = draw_fractal(n, m_barnsley, p_barnsley)
ax.scatter(barnsley[:,0], barnsley[:,1], s=0.006, marker='.', color=colors[1][0])
plt.savefig('../results/plots/03_barnsley_fern_ifs.png', dpi=1200)
plt.show()
plt.close()
```

```python
r = range(14)
t_0 = 8

def dimensions(fractal_list):
  fraction = []
  for i in range(len(r)):
    histogram = np.histogram2d(fractal_list[:,0], fractal_list[:,1], bins=2**r[i])[0]
    fraction.append(np.where(histogram == 0, 0, 1).sum())
  return fraction

def model(x, a):
  return a * x

fig, ax = plt.subplots()
points_barnsley = np.log(dimensions(barnsley))
points_sierp = np.log(dimensions(sierpinski))

ax.scatter(r, points_sierp, marker='.', color=colors[0][0])
ax.scatter(r, points_barnsley, marker='.', color=colors[1][0])

popt, pcov = curve_fit(model, r[:t_0], points_sierp[:t_0], p0=[1])
ax.plot(r[:t_0], model(r[:t_0], *popt), label=f'Sierpinski a = {popt[0]:.3f}', zorder=1, color=colors[0][1])



popt, pcov = curve_fit(model, r[:t_0], points_barnsley[:t_0], p0=[1])
ax.plot(r[:t_0], model(r[:t_0], *popt), label=f'Barnsley a = {popt[0]:.3f}', zorder=1, color=colors[1][1])

ax.legend()
plt.show()
plt.close()
```
