```python
import numpy as np
from scipy.integrate import solve_ivp
from matplotlib.widgets import Button, Slider
import matplotlib.pyplot as plt
from numpy.random import random
```

```python
colors = [['dodgerblue', 'indigo'],
          ['orangered', 'sienna'],
          ['gold', 'olive']]

amplitude_0 = 0.4
x_0, v_0 = 0, 0.15
t_0 = 1000

a, b = 0, 200
t = np.linspace(a, b, 10000)

def duffing(t, xv):
  x, v = xv
  f = amplitude_0
  b, a, c, omega = 1, 1, 0.2, 0.213 * 2 * np.pi
  return [v, b * x - a * x**3 - c * v + f * np.cos(omega * t)]

def equilibrium(t, xv):
  x, v = xv
  b, a, c, omega = 1, 1, 0.2, 0.213 * 2 * np.pi
  return [v, b * x - a * x**3 - c * v]

sol = solve_ivp(duffing, [a, b], [x_0, v_0], t_eval=t)
equ = solve_ivp(equilibrium, [a, b], [x_0, v_0], t_eval=t)
```

```python
fig, ax = plt.subplots()
fig.suptitle(f'Duffing System $f = {amplitude_0}$')
ax.set_xlabel("$t$")
ax.annotate("$x(t)$", xy=(-0.15, 0.55), xycoords='axes fraction', color=colors[0][0])
ax.annotate("$v(t)$", xy=(-0.15, 0.45), xycoords='axes fraction', color=colors[1][0])

ax.set_xlim(0, 75)
ax.plot(sol.t, sol.y[0], color=colors[0][0])
ax.plot(sol.t, sol.y[1], color=colors[1][0])
ax.plot(equ.t, [equ.y[0, -1]] * len(t), '--', color=colors[0][1])
ax.plot(equ.t, [equ.y[1, -1]] * len(t), '--', color=colors[1][1])
plt.show()
plt.close()
```

```python
def equ_points():
    points = []
    for _ in range(1, 10):
      equ = solve_ivp(equilibrium, [a, b], [x_0 + random(), 10 * v_0 * random()], t_eval=t)
      points.append(equ.y[:, -1])
    return points

points = equ_points()
x_values = [point[0] for point in points]
y_values = [point[1] for point in points]

fig, ax = plt.subplots()
fig.suptitle(f'Duffing System Phase Space $f = {amplitude_0}$')
ax.set_xlabel("$x(t)$")
ax.set_ylabel("$v(t)$")
ax.plot(sol.y[0][t_0:], sol.y[1][t_0:], color=colors[2][0])
ax.plot(x_values, y_values, 'o', color=colors[2][1], label='Equilibrium Points')
ax.legend()

plt.show()
plt.close()
```

```python
amplitude_0_new = 0.3
t_new = np.linspace(a, b, int(((b - a) + 1) / 0.213))

def duf(t, xv):
  x, v = xv
  f = amplitude_0_new
  b, a, c, omega = 1, 1, 0.05, 0.213 * 2 * np.pi
  return [v, b * x - a * x**3 - c * v + f * np.cos(omega * t)]

sol_new = solve_ivp(duf, [a, b], [x_0, v_0], t_eval=t_new)
xsol, vsol = sol_new.y
fig, ax = plt.subplots()
fig.suptitle(f'Duffing System Poincare Section $f = {amplitude_0_new}$')
ax.set_xlabel("$x(t)$")
ax.set_ylabel("$v(t)$")
plt.scatter(sol_new.y[0], sol_new.y[1], s=3, c='b', lw=0, marker='o', label='punkty')

plt.show()
```
