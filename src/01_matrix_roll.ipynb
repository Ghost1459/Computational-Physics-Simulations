```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import PillowWriter, FuncAnimation
```

```python
n = 64
matrix = np.random.random((n,n))

fig, ax = plt.subplots()
image = ax.imshow(matrix, cmap='viridis', interpolation='nearest')
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
ax.axis('off')

def update(frames):
  rolled = np.roll(matrix, frames, axis=0)
  image.set_array(rolled)
  return [image]

anim = FuncAnimation(fig, update, frames=n, blit=True, repeat=True)
anim.save("../results/gifs/01_matrix_roll.gif", writer=PillowWriter(fps=30))
```
